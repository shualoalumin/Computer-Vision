```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 기본 URL 설정
base_url = "https://dics.co.kr/bbs/board.php?bo_table=dinews&page="

# 결과를 저장할 리스트
admission_data = []

# User-Agent 헤더 추가
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# 페이지 범위 설정 (1-21페이지까지)
for page_num in range(1, 22):
    url = f"{base_url}{page_num}"
    response = requests.get(url, headers=headers)
    
    print(f"페이지 {page_num} 처리 중...")  # 디버깅을 위한 출력
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        # 게시물 리스트 선택 (수정된 선택자)
        articles = soup.select("td.list_subject a")
        
        print(f"페이지 {page_num}에서 찾은 게시물 수: {len(articles)}")  # 디버깅을 위한 출력
        
        for article in articles:
            title = article.text.strip()
            link = article.get("href")
            if link and not link.startswith('http'):
                link = 'https://dics.co.kr' + link
            
            # 진학 정보만 선택
            if "진학" in title or "장학금" in title:
                admission_data.append({
                    "Title": title,
                    "Link": link
                })
                print(f"저장된 게시물: {title}")  # 디버깅을 위한 출력
    else:
        print(f"페이지 {page_num}에 접근할 수 없습니다. 상태 코드: {response.status_code}")

print(f"총 저장된 게시물 수: {len(admission_data)}")  # 디버깅을 위한 출력

# DataFrame으로 변환 후 엑셀로 저장
df = pd.DataFrame(admission_data)
df.to_excel("DICS_admission_results.xlsx", index=False, engine='openpyxl')

print("크롤링 완료 및 엑셀 저장 완료.")
```
